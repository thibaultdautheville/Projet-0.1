## Synthèse générale


Nous disposons :

- De 241 pays / régions  
- De 3665 indicateurs uniques  
- De données couvrant la période 1970–2100  
- D’un total de 886 930 combinaisons pays × indicateurs  

La structure globale correspond à un modèle en étoile (star schema) :

StatsSeries5 (Dimension Indicateur)  
        ↑  
StatsCountry2 ← StatsData3 → StatsFootNote4  
        ↓  
CountrySeries1  

StatsData3 est la table centrale autour de laquelle s’articulent les dimensions pays, indicateurs et notes.


```python

#Récupération des noms de colonnes et des dimensions de chaque DataFrame

names = [
    "CountrySeries1","StatsCountry2", "StatsData3", "StatsFootNote4", "StatsSeries5",
]

for name in names:
    df = globals().get(name)
    if isinstance(df, pd.DataFrame):
        print(f"{name} — shape: {df.shape}")
        print(df.columns.tolist())
        print()
    else:
        print(f"{name} introuvable ou n'est pas un DataFrame.\n")
```

```python
# Faire apparaître les deux premières lignes de chaque DataFrame

datasets = {
    'CountrySeries1': CountrySeries1,
    'StatsCountry2': StatsCountry2,
    'StatsData3': StatsData3,
    'StatsFootNote4': StatsFootNote4,
    'StatsSeries5': StatsSeries5
}

for dataset_name, df in datasets.items():
    print(f"\n{'='*80}")
    print(f"DATASET: {dataset_name}")
    print(f"{'='*80}\n")
    
    for row_idx in range(min(2, len(df))):
        print(f"\n--- Row {row_idx} ---")
        for col_name in df.columns:
            cell_value = df.loc[row_idx, col_name]
            print(f"{col_name}: {cell_value}")
```
